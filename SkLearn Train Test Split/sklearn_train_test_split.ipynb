# Scikit-learn Basics: Train-Test Split

## Why Train-Test Split?
We train models on known data and test on unseen data to avoid overfitting.

```python

from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Sample dataset
data = {
    'Hours_Studied': [1,2,3,4,5,6,7,8],
    'Marks': [10,20,30,40,50,60,70,80]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['Hours_Studied']]
y = df['Marks']
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# class Test_size means - Kitna data test ke liye rakhna hai
# class random_state means - Data ko shuffle karne ke liye ek seed value dena hai, taki har baar same data shuffle ho
# class X_train, X_test, y_train, y_test means - Train aur test data ko alag karna hai, X_train aur y_train model ko train karne ke liye use hoga, X_test aur y_test model ko evaluate karne ke liye use hoga
```

```python
X_train, X_test, y_train, y_test
```

## Key Points
- test_size defines how much data is reserved for testing
- random_state ensures reproducibility
